```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

# Target for SarsCOV

```python
target = new_client.target
query_target = target.search("coronavirus")
query_target_df = pd.DataFrame.from_dict(query_target)
query_target_df
```

```python
selected_target = query_target_df.target_chembl_id[5]
selected_target
```

# Taking CHEMBL4296578 reported as IC50

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
res_df = pd.DataFrame.from_dict(res)
res_df
```

```python
res_df.head(3)
```

```python
res_df.standard_type.unique()
```

```python
res_df.to_csv('sarscov_bioactivity_data.csv', index=False)
```

```python
sarscov_bioactivity_data.csv
```

```python
res_df_updated = res_df[res_df.standard_value.notna()]
res_df_updated
```

# Performing data preprocessing
    - Lableling compounds using active, inactive and intermediate

Active - considered when compound value is < 1000nM
Inactive - considered when compound value is > 1000nM
Intermediate - considered when compound value is in between 1000nM - 1000nM

```python
bioactivity_class = []
for i in res_df_updated.standard_value:
    if float(i) > 1000:
        bioactivity_class.append("inactive")
    elif float(i) < 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

chembl_id Iteration

```python
mol_cid = []
for i in res_df_updated.molecule_chembl_id:
    mol_cid.append(i)
```

canconical_smiles

```python
cano_sm = []
for i in res_df_updated.canonical_smiles:
    cano_sm.append(i)
```

Standard Values

```python
std_val = []
for i in res_df_updated.standard_value:
    std_val.append(i)
```

Combining all 4

```python
data_tuples = list(zip(mol_cid, cano_sm, bioactivity_class, std_val))
res_df_final = pd.DataFrame (data_tuples, columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```

```python
res_df_final.to_csv('preprocessed_bioactivity_data_csv.csv', index=False)
```
